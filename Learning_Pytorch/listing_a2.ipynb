# Seeing models as computational graphs

## A Logistic Regression forward pass

```python
import torch
import torch.nn.functional as F

# 1. Initialize tensors
y = torch.tensor([1.0])
x1 = torch.tensor([1.1])
w1 = torch.tensor([2.2])
b = torch.tensor([0.0])

# 2. Forward pass (Linear combination)
z = x1 * w1 + b  # 1.1 * 2.2 + 0 = 2.42

# 3. Activation function (Sigmoid)
a = torch.sigmoid(z)  # 1 / (1 + e^-2.42) ≈ 0.9183

# 4. Compute Loss
loss = F.binary_cross_entropy(a, y)

print(f"Prediction (a): {a.item():.4f}")
print(f"Loss: {loss.item():.4f}")
```
