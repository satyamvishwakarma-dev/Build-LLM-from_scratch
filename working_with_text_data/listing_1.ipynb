### Testing the tokenization of a text

```python
import re
text = "Hello, world. This is a test."
result = re.split(r'(\s)', text)
print(result)
```

#### Modify the above code to split everything

```python
import re
text = "Hello, world. This is a test."
result = re.split(r'([,.]|\s)', text)
print(result)
```
